```python
class bank:
    bankname='canara bank'
    branch='Gtb nagar,india'
    
    #create account 
    def __init__(self,username,pan,address):
        self.username=username
        self.pan=pan
        self.address=address
        self.balance=0.0
        print(f'hello {self.username} congratulation! your bank account created succesfully')
        
    #Deposit
    def deposit(self,amount):
        
        self.balance=self.balance+amount
        print(f'{amount} deposited successfully')
    
    #withdraw
    def withdraw(self,amount):
        if amount<self.balance:
            self.balance-=amount
            print(f'{amount} withdraw successfully, current balance {self.balance}')
        else:
            print('insfuicient fund')
            
    #ministatemnt
    def ministatement(self):
        print(f'Account Holder: {self.username}')
        print(f'PAN: {self.pan}')
        print(f'Address: {self.address}')
        print(f'Current Balance: {self.balance}')
     
    #Stop
    
    
print(f'Welcome to {bank.bankname},{bank.branch}')

username=input('enter the username:-')
pan=input('enter the pan:-')
address=input('enter the address:-')


b=bank(username,pan,address)

while True:
    print('\n please select any option :-')
    
    print('1.Deposit\n2.Withdraw\n3.ministatment\n4.stop\n')
    option =int(input(''))
    
    if option==1:
        amount=float(input('enter the Deposit amount'))
        b.deposit(amount)
    elif option == 2:
        amount = float(input('Enter the withdrawal amount: '))
        b.withdraw(amount)
    elif option == 3:
        b.ministatement()    
    elif option==4:
        print('Thanks for using canara bank')
        break
    else:
        print('invalid option. please try again')
    






    
```
